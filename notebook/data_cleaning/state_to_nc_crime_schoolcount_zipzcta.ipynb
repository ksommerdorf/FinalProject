### nanda school count by zcta 2000-2018, zcta is reference to zipcode

```python
school_df = pd.read_csv("nanda_schoolct_zcta_2000-2018_01P.csv")
school_df.head(5)
```

```python
school_df.rename(columns = {'zcta19': 'ZCTA'}, inplace=True)
```

```python
school_reduce_df = school_df[["ZCTA", "year", "all_school"]]
school_reduce_df.head()
```

```python
nc_check_df = school_reduce_df.loc[school_reduce_df["ZCTA"] == 27019]
nc_check_df.head(5)
```

### zip to zcta csv filter to join with nanda school count by ZCTA

```python
zipToZcta_df = pd.read_csv("ZiptoZcta_Crosswalk_2021.csv")
zipToZcta_df.head(5)
```

```python
zipToZctaUpdate_df = pd.read_csv("ZiptoZctaUpdate2022.csv")
zipToZctaUpdate_df.head(5)
# csv file contained two sheets, this is 2nd read in to join with zipTOZcta_df
```

```python
frames = [zipToZcta_df, zipToZctaUpdate_df]

entire_zipToZcta_df = pd.concat(frames)
```

```python
entire_zipToZcta_df.head()
```

```python
len(entire_zipToZcta_df)
```

```python
entire_zipToZcta_df.drop('ZIP_TYPE', axis=1, inplace=True)
```

```python
entire_zipToZcta_df.head()
```

```python
nc_zip_to_zcta_df = entire_zipToZcta_df[entire_zipToZcta_df["STATE"] == "NC"]
```

```python
nc_zip_to_zcta_df.head()
```

#### check value types for merge, ZCTA needs changing in nc zip zcta school df

```python
school_reduce_df.dtypes
```

```python
nc_zip_to_zcta_df.dtypes
```

```python
nc_zip_to_zcta_df["ZCTA"] = nc_zip_to_zcta_df["ZCTA"].astype(np.int64)
print(nc_zip_to_zcta_df.dtypes)
```

```python
# join using merge inner
merge_nc_zip_zcta_school_df = pd.merge(nc_zip_to_zcta_df, school_reduce_df, on="ZCTA", how="inner")
```
