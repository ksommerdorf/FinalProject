```python
import numpy as np
import pandas as pd
from pathlib import Path
from collections import Counter
```

```python
file_path = Path('RDC_Inventory_Core_Metrics_State_History.csv')
df = pd.read_csv(file_path)
df.head(50)
```

```python
df.dtypes
```

```python
# retrieve all rows that state = nc
north_carolina_df = df.loc[df['state_id'] == 'nc']
north_carolina_df
```

```python
# NaNs status
north_carolina_df.isnull().sum()
```

```python
len(north_carolina_df)
```

```python
nc_drop_df = north_carolina_df.dropna()
nc_drop_df.isnull().sum()
```

```python
len(nc_drop_df)
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

plt.rcParams["figure.figsize"] = (30,8)

# plot
fig, ax = plt.subplots()
x = nc_drop_df["month_date_yyyymm"]
y = nc_drop_df["median_listing_price"]


ax.plot(x, y, linewidth=2.0)

ax.set()

plt.show()
```

```python
# plot2 graph for inverted axes


plt.rcParams["figure.figsize"] = (30,8)

# plot
fig, ax = plt.subplots()
x = nc_drop_df["month_date_yyyymm"]
y = nc_drop_df["median_listing_price"]


ax.plot(x, y, linewidth=2.0)
ax.invert_xaxis()
ax.set()

plt.show()
```
